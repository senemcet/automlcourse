```python
# Sayısal sütunları belirle
numerical_cols = ['age', 'avg_glucose_level', 'bmi', 'hypertension', 'heart_disease']

print("SAYISAL DEĞİŞKENLER İÇİN İSTATİSTİKLER:")
print("-"*70)

stats_summary = pd.DataFrame()
for col in numerical_cols:
    stats_summary[col] = {
        'Ortalama': df[col].mean(),
        'Medyan': df[col].median(),
        'Standart Sapma': df[col].std(),
        'Minimum': df[col].min(),
        'Maksimum': df[col].max(),
        'Eksik Değer': df[col].isnull().sum(),
        'Eksik Değer %': (df[col].isnull().sum() / len(df)) * 100
    }

stats_summary = stats_summary.T
print(stats_summary.round(2))
```

```python
# 2. EKSİK DEĞER ANALİZİ
# ============================================
print("\n" + "="*70)
print(" 2. EKSİK DEĞER ANALİZİ")
print("="*70 + "\n")

missing_data = pd.DataFrame({
    'Sütun': df.columns,
    'Eksik Değer Sayısı': df.isnull().sum().values,
    'Eksik Değer %': (df.isnull().sum().values / len(df) * 100).round(2),
    'Veri Tipi': df.dtypes.values
})

print(missing_data)
print(f"\nToplam Satır Sayısı: {len(df)}")
print(f"Toplam Eksik Değer: {df.isnull().sum().sum()}")
```

```python
# 4. GÖRSELLEŞTİRMELER
# ============================================

# 4.1 Hedef Dağılımı (Stroke)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Hedef Değişken Dağılımı (Stroke)', fontsize=16, fontweight='bold')

# Bar plot
stroke_counts = df['stroke'].value_counts()
colors = ['#2ecc71', '#e74c3c']
axes[0].bar(stroke_counts.index, stroke_counts.values, color=colors, alpha=0.7, edgecolor='black')
axes[0].set_xlabel('Stroke (0: Hayır, 1: Evet)', fontsize=12)
axes[0].set_ylabel('Frekans', fontsize=12)
axes[0].set_title('Stroke Dağılımı', fontsize=13)
axes[0].set_xticks([0, 1])
for i, v in enumerate(stroke_counts.values):
    axes[0].text(i, v + 50, str(v), ha='center', fontweight='bold')

# Pie chart
axes[1].pie(stroke_counts.values, labels=['Stroke Yok (0)', 'Stroke Var (1)'],
            autopct='%1.1f%%', colors=colors, startangle=90,
            explode=(0.05, 0.05), shadow=True)
axes[1].set_title('Stroke Oranı', fontsize=13)

plt.tight_layout()
plt.show()
```

```python
# 4.3 Aykırı Değerler (Boxplot)
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Aykırı Değer Analizi (Boxplot)', fontsize=16, fontweight='bold')

# Age
axes[0, 0].boxplot(df['age'].dropna(), vert=True, patch_artist=True,
                    boxprops=dict(facecolor='lightblue', alpha=0.7),
                    medianprops=dict(color='red', linewidth=2))
axes[0, 0].set_ylabel('Yaş', fontsize=11)
axes[0, 0].set_title('Yaş Dağılımı', fontsize=12)
axes[0, 0].grid(True, alpha=0.3)

# Average Glucose Level
axes[0, 1].boxplot(df['avg_glucose_level'].dropna(), vert=True, patch_artist=True,
                    boxprops=dict(facecolor='lightgreen', alpha=0.7),
                    medianprops=dict(color='red', linewidth=2))
axes[0, 1].set_ylabel('Ortalama Glukoz Seviyesi', fontsize=11)
axes[0, 1].set_title('Glukoz Seviyesi Dağılımı', fontsize=12)
axes[0, 1].grid(True, alpha=0.3)

# BMI
axes[1, 0].boxplot(df['bmi'].dropna(), vert=True, patch_artist=True,
                    boxprops=dict(facecolor='lightcoral', alpha=0.7),
                    medianprops=dict(color='red', linewidth=2))
axes[1, 0].set_ylabel('BMI', fontsize=11)
axes[1, 0].set_title('BMI Dağılımı', fontsize=12)
axes[1, 0].grid(True, alpha=0.3)

# Yaş vs Stroke
stroke_yes = df[df['stroke'] == 1]['age']
stroke_no = df[df['stroke'] == 0]['age']
bp = axes[1, 1].boxplot([stroke_no, stroke_yes], labels=['Stroke Yok', 'Stroke Var'],
                         patch_artist=True,
                         boxprops=dict(alpha=0.7),
                         medianprops=dict(color='red', linewidth=2))
for patch, color in zip(bp['boxes'], ['lightblue', 'orange']):
    patch.set_facecolor(color)
axes[1, 1].set_ylabel('Yaş', fontsize=11)
axes[1, 1].set_title('Stroke Durumuna Göre Yaş', fontsize=12)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 4.4 Aykırı değer istatistikleri
print("\n" + "="*70)
print(" 5. AYKIRI DEĞER İSTATİSTİKLERİ (IQR Yöntemi)")
print("="*70 + "\n")

for col in ['age', 'avg_glucose_level', 'bmi']:
    data = df[col].dropna()
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]

    print(f"{col.upper()}:")
    print(f"  Q1 (25%): {Q1:.2f}")
    print(f"  Q3 (75%): {Q3:.2f}")
    print(f"  IQR: {IQR:.2f}")
    print(f"  Alt Sınır: {lower_bound:.2f}")
    print(f"  Üst Sınır: {upper_bound:.2f}")
    print(f"  Aykırı Değer Sayısı: {len(outliers)} ({len(outliers)/len(data)*100:.2f}%)")
    print("-"*70)
```
