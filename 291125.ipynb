```python
import sklearn
```

```python
#gridsearchCV-gerekli kutuphaneler

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

#veriseti yukle
cancer= load_breast_cancer()
X,y=cancer.data,cancer.target
print(X,y)
```

```python
pipeline=Pipeline([
    ('scaler',StandardScaler()),
    ('model',LogisticRegression(max_iter=2000))
])

param_grid=[
    {
        #L2 regularizasyonu için parametreler
        "model__solver":["lbfgs","liblinear"],
        "model__penalty":["l2"],
        "model__C":[0.001, 0.01, 0.1,1,10,100]
    },
    {
        #L1 regularizasyonu için parametreler
        "model__solver":["liblinear","saga"],
        "model__penalty":["l1"],
        "model__C":[0.01, 0.1,1,10]
    }
]
```

```python
#GridSearchCV tanimlamasi
grid=GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1  #var olan tum cekirdekleri kullan
)

grid.fit(X,y)  #gridsearch ile en iyi parametreleri bul
```

```python
# Sonuclari yazdir
print("en iyi parametreler:",grid.best_params_)
print("en iyi skor:", grid.best_score_)
print("en iyi model:", grid.best_estimator_)

```

```python
#RandomizedSearchCV

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from scipy.stats import loguniform

#veriseti yukle
cancer= load_breast_cancer()
X,y=cancer.data,cancer.target
print(X,y)
```

```python
pipeline=Pipeline([
    ('scaler',StandardScaler()),
    ('model',LogisticRegression(max_iter=2000))
])

#parametre dagilimi
param_dist={
    "model__C":loguniform(1e-3, 1e3),
    "model__penalty":["l2"],
    "model__solver":["lbfgs"]
}

# RandomizedSearchCV tanimlamasi
random_search=RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=20,
    cv=5,
    scoring="accuracy",
    verbose=10,
    n_jobs=-1,
    random_state=42
)

random_search.fit(X,y)

#sonuclar

import pandas as pd
results=pd.DataFrame(random_search.cv_results_)
results[["params","mean_test_score","rank_test_score"]].sort_values(by="rank_test_score").head()
```
