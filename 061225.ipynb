```python
#onehot-label-ordinal encoder ornekleri
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder
```

```python
veri = np.array([
    ['Kırmızı', 'Küçük', 'Lise'],
    ['Mavi', 'Orta', 'İlkokul'],
    ['Yeşil', 'Büyük', 'Üniversite'],
    ['Kırmızı', 'Küçük', 'Lise']
])
```

```python
# OneHotEncoder örneği (renkler için ayrı sütunlar açıyoruz)
onehot_encoder = OneHotEncoder(sparse_output=False)
renk_encoded = onehot_encoder.fit_transform(veri[:, 0].reshape(-1, 1))
```

```python
# LabelEncoder örneği ( sıralama)
label_encoder = LabelEncoder()
boyut_encoded = label_encoder.fit_transform(veri[:, 1])
```

```python
# OrdinalEncoder örneği (kendi içinde hiyerarşisi olacak şekilde sıralı şekilde sıralıyoruz)
ordinal_encoder = OrdinalEncoder(categories=[['İlkokul', 'Lise', 'Üniversite']])
egitim_encoded = ordinal_encoder.fit_transform(veri[:, 2].reshape(-1, 1))
```

```python
# Sonuçları birleştirme
sonuc = np.hstack((renk_encoded, boyut_encoded.reshape(-1, 1), egitim_encoded))
```

```python
# Orijinal verileri ve dönüştürülmüş verileri yazdırma
print("Orijinal Veri:\n", veri)
print("\nDönüştürülmüş Veri:\n", sonuc)
```

```python
# Her encoder’ın kategorilerini ve dönüştürülmüş değerlerini yazdırma
print("\nOneHotEncoder Kategorileri:\n", onehot_encoder.categories_)
print("\nLabelEncoder Kategorileri:\n", label_encoder.classes_)
print("\nOrdinalEncoder Kategorileri:\n", ordinal_encoder.categories_)
```
